#### Starter Code for Question 1

```python
# Definition for a binary tree node.
class TreeNode(object):  
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def is_duplicate(root: TreeNode) -> int:
    if not root:
        return -1  # Return -1 if the tree is empty

    seen = set()
    queue = [root]  # Use a queue for BFS (level-order traversal)

    while queue:
        node = queue.pop(0)  # Process the first node in the queue
        
        if node.val in seen:
            return node.val  # Return the first duplicate found
        seen.add(node.val)

        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

    return -1  # No duplicate found

```


-   Explain why your solution works


```python
# My solution works by using Breadth-First Search (BFS), which means traversing the tree level by level using a queue. I also keep track of values that have already been seen in a set to allow for fast lookups (O(1)).

# As I go through the tree:
# I process each node one by one.
# If the node's value is already in the set, I return it immediately as the first duplicate. Since BFS processes nodes closest to the root first, this ensures the first duplicate found is the one closest to the root.
# If the node’s value is not in the set, I add it to the set and continue checking its children.
# If I finish traversing the entire tree and no duplicates are found, I return -1.
```


-   Explain the problem’s time and space complexity


```python
# The time complexity is O(N) because the solution visits each node exactly once using 
# Breadth-First Search (BFS). Since we process every node only once and perform constant-time 
# (O(1)) operations like checking values in a set and adding nodes to a queue, the overall 
# complexity remains O(N), where N is the number of nodes in the tree.

# The space complexity is also O(N) because we store visited values in a set (seen) and use a 
# queue (queue) to process nodes level by level. In the worst case, where the tree is balanced, 
# the queue could store up to N/2 nodes at the last level, leading to O(N) space usage.
```

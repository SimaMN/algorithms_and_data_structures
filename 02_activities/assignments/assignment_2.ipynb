-   Paraphrase the problem in your own words.


The problem involves designing a function that, given the root of a binary tree, returns all possible paths from the root to the leaf nodes. Each path should be represented as a sequence of node values, starting from the root and ending at a leaf. The order of the paths in the result is not important. 


-   Create 1 new example that demonstrates you understand the problem. Trace/walkthrough 1 example that your partner made and explain it.


Example 1 :

Input: A binary tree represented as [10, 5, 15, 3, 7, null, 20].

Output: [[10, 5, 3], [10, 5, 7], [10, 15, 20]]

The partner's example :

Input: root = [5, 3, 8, 2, 4, None, 9]

Output: [[5, 3, 2], [5, 3, 4], [5, 8, 9]]

Walkthrough the example:

Start at root: {5}
Extend to left: {5, 3}
Extend to left: {5, 3, 2} → leaf reached, record [5, 3, 2]
Backtrack to: {5, 3}
Extend to right: {5, 3, 4} → leaf reached, record [5, 3, 4]
Backtrack to: {5}
Extend to right: {5, 8}
Extend to right: {5, 8, 9} → leaf reached, record [5, 8, 9]
Final output: [[5, 3, 2], [5, 3, 4], [5, 8, 9]]


-   Copy the solution your partner wrote. 


```python
# Definition for a binary tree node.
class TreeNode(object):
     def __init__(self, val = 0, left = None, right = None):
         self.val = val
         self.left = left
         self.right = right
def bt_path(root: TreeNode) -> List[List[int]]:


    if not root:
        return []

    result = []

    def bt_path_helper(node, path):
        path.append(node.val)

        if not node.left and not node.right:  # If it's a leaf node
            result.append(path[:])  # Append a copy of the path

        if node.left:
            bt_path_helper(node.left, path)

        if node.right:
            bt_path_helper(node.right, path)

        path.pop()  # remove the last node to reset the path for the next branch

    bt_path_helper(root, [])
    return result
```


-   Explain why their solution works in your own words.


Their solution works by using a recursive depth-first search (DFS) approach combined with backtracking to explore all possible root-to-leaf paths. Here's a breakdown of why it works:

+ Recursive Traversal: The helper function is called recursively for each node, ensuring that every branch of the tree is explored.

+ Path Building: As the recursion descends into the tree, the current node's value is appended to the ongoing path. This builds up the path from the root to the current node.

+ Leaf Detection: When the function encounters a leaf node (a node with no left or right children), it recognizes that a complete root-to-leaf path has been formed. A copy of this path is then added to the result list.

+ Backtracking: After processing a node's children, the function removes the node's value from the path before returning. This step is crucial because it resets the path, allowing the correct path to be built for other branches of the tree.

+ Complete Exploration: By recursively calling itself for both the left and right children (if they exist) and by backtracking after each recursive call, the function ensures that all possible paths from the root to every leaf are recorded.

In summary, the solution correctly explores the tree, builds each path step-by-step, and uses backtracking to ensure that each branch's path is isolated and accurately captured.


-   Explain the problem’s time and space complexity in your own words.


+ Time Complexity:

Each node is visited once, but at each leaf, a copy of the current path (of length O(H)) is created.
In the worst-case (unbalanced tree with H = O(N)), this results in O(N²) time; for a balanced tree (H = O(log N)), it's O(N log N).

+ Space Complexity:

The recursion stack uses O(H) space.
Additionally, storing all paths requires O(L * H) space (with L being the number of leaves), which in the worst-case can be O(N * H).


-   Critique your partner's solution, including explanation, and if there is anything that should be adjusted.


While the code is overall correct and well-structured, here are a few minor points that could be adjusted:

+ Alternative Copy Method:
Using path.copy() instead of slicing (i.e., path[:]) may improve readability by clearly indicating that a copy of the list is being made.

+ Recursion Depth:
For very deep trees, a recursive approach might hit recursion limits. An iterative solution or tail recursion optimization (if applicable) could be considered, though this is typically not a problem for most practical inputs.
